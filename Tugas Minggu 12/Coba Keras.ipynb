```python
import tensorflow
from tensorflow import keras
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from matplotlib import pyplot
from keras.callbacks import EarlyStopping
import time
```

```python
#Loading Data Into Data Frame

df = pd.read_csv("data.csv") 
df.head(5)                                 # Display the top 5 rows
```

```python
df.count()
```

```python
df = df.drop_duplicates()      #Drop duplicate row
```

```python
df.count()
```

```python
#Dropping the missing and null values

print(df.isnull().sum())         #count the number of missing and null values
```

```python
df = df.dropna()    #Drop the missing values.
df.count()          #Count the number of rows after dropping the missing and null values
```

```python
print(df.isnull().sum())   #Check if there are some still missing or null values left
```

```python
df['MSRP'] = (df['MSRP'] >= 40000).astype(int)
df.head(5)
```

```python
for col in df.columns:
    df = df[df[col].notnull()]
```

```python
string_col = ['Make', 'Model', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Market Category', 'Vehicle Size', 'Vehicle Style']
for col in string_col:
    df[col] = df[col].astype('category')
    df[col] = df[col].cat.codes

df.head(5)
```

```python
feature_cols = ['Make', 'Model', 'Year', 'Engine Fuel Type','Engine HP','Engine Cylinders','Transmission Type', 'Driven_Wheels', 'Vehicle Size', 'Vehicle Style', 'highway MPG', 'city mpg', 'Popularity']
X = df[feature_cols] # Features
y = df.MSRP # Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
```

```python
X_train
```

```python
X_test
```

```python
model=keras.models.Sequential([
    keras.layers.Dense(64, input_dim = X_train.shape[1], activation='tanh'),  
    keras.layers.Dropout(0.3),
    
    keras.layers.Dense(units=32, activation='tanh'), 
    keras.layers.Dropout(0.2),
    
    keras.layers.Dense(units=1, activation="sigmoid"),
],name="Initial_model",)
model.summary()
```

```python
learning_rate = 0.01
optimizer = keras.optimizers.Adam(lr=learning_rate)
```

```python
model.compile(optimizer=optimizer,
            loss='binary_crossentropy', metrics=['accuracy'])
```

```python
early_stopping_monitor = EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=100,
    verbose=0,
    mode='auto',
    baseline=None,
    restore_best_weights=True
)
```

```python
start = time.clock() 
history = model.fit(X_train, y_train, epochs=1000, batch_size=512, validation_data=(X_test, y_test), verbose=1, callbacks=[early_stopping_monitor])
end = time.clock()
```

```python
print("Train time: {} ".format(end-start))
```

```python
predicted = model.predict(X_test)
```
